Introduction to simple construction of neural networks with keras

* Constructing simple feed forward neural network

* Constructing convolutional neural network model

* Compiling models

```python
import tensorflow as tf
```

```python
print(tf.__version__)
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Flatten, Softmax, Dense, Conv2D, MaxPooling2D
```

Building a feedforward neural network


```python
model = Sequential([
    Flatten(input_shape=(28,28)),
    Dense(16, activation='relu'),
    Dense(16, activation='relu'),
    Dense(10, activation='softmax')
])
```

```python
model.weights
```

```python
model.summary()
```

Building a convolutional model

```python
conv_model = Sequential([
          Conv2D(16, (3, 3), activation='relu', input_shape=(32, 32, 3)),
          MaxPooling2D((3, 3)),
          Flatten(),
          Dense(64, activation='relu'),
          Dense(10, activation='softmax')
])
```

```python
conv_model.summary()
```

Compile the model


```python
conv_model.compile(optimizer='adam',
                   loss='sparse_categorical_crossentropy',
                   metrics=['accuracy'])
```

```python
print(conv_model.loss)
print(conv_model.optimizer)
print(conv_model.metrics)
```

```python
print(conv_model.optimizer.lr)
```
